## Seaborn
[Quickstart](https://seaborn.pydata.org/tutorial/introduction.html)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/df_filled.csv', parse_dates = ['date'])
df.head(2)
```

```python
sns.histplot(df['co'], bins=30, kde=True)

plt.title('Histogram - Seaborn')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
sns.scatterplot(x=df['pm10'], y=df['pm25'])

plt.title('Scatter Plot - Seaborn')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```
