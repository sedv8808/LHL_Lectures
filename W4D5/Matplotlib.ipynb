## Matplotlib

[Quickstart](https://matplotlib.org/stable/users/explain/quick_start.html)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/df_filled.csv', parse_dates = ['date'])
df.head()
```

```python
plt.hist(df['co'], bins=30, alpha=0.7)

plt.title('Histogram with Matplotlib')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```

```python
plt.scatter(df['pm10'], df['pm25'], c='blue')

plt.title('Scatter Plot - Matplotlib')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```
